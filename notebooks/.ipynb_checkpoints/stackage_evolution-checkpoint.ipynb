### Utilidades

```python
# retorna una lista con todos los paquetes existentes en alguna version de LTS
def get_all_time_packages():
    all_packages = set()
    for df in df_list:
        lts_packages = set(df['package'])
        all_packages = all_packages.union(lts_packages)

    return list(all_packages)
```

```python
all_time_packages = get_all_time_packages()
```

```python
# construye un dataframe que indica la continuidad de los paquetes recibidos por parametro en las versiones LTS
def build_continuity_matrix(index):
    df = pd.DataFrame(index=index, columns=lts_list)

    for i, row in df.iterrows():
        for j, lts in enumerate(lts_list):
            flag = 0
            if(not df_list[j].loc[df_list[j]['package'].isin([row.name])].empty):
                flag = 1
            df.at[row.name, lts] = float(flag)

    for idx, lts in enumerate(lts_list):
            df[lts] = pd.to_numeric(df[lts]) 
            
    return df
```

```python
continuity_df = build_continuity_matrix(get_all_time_packages())
```

### Graficar la cantidad de paquetes que se agregaron que usan MTL en cada version LTS

```python
def build_mtl_continuity_matrix(index):
    df = pd.DataFrame(index=index, columns=lts_list)

    for i, row in df.iterrows():
        for j, lts in enumerate(lts_list):
            state = 0
            package = df_list[j].loc[df_list[j]['package'].isin([row.name])]

            if(not package.empty):
                state = 1
                [use_mtl] = package['mtl-direct'].values
                if(use_mtl == 1):
                    state = 2
            
            df.at[row.name, lts] = float(state)

    for idx, lts in enumerate(lts_list):
            df[lts] = pd.to_numeric(df[lts]) 
            
    return df
```

```python
mtl_continuity_df = build_mtl_continuity_matrix(get_all_time_packages())
```

```python
plt.figure(figsize=(18,6))
ax = sns.heatmap(mtl_continuity_df, xticklabels=1, yticklabels=False)
```

```python
def get_added_packages_mtl_by_lts(continuity_df):
    count = list(np.zeros(len(lts_list)))

    for i, row in continuity_df.iterrows():
        for j, lts in enumerate(lts_list):
            if(row[lts] == 2):
                if(j == 0):
                    count[j] += 1
                elif(row[j-1] < 2):
                    count[j] += 1
    
    return count

def get_removed_packages_mtl_by_lts(continuity_df):
    count = list(np.zeros(len(lts_list)))
    
    for i, row in continuity_df.iterrows():
        for j, lts in enumerate(lts_list):
            if(j > 0):
                if((row[j] == 1 or row[j] == 0) and row[j-1] == 2):
                    count[j] += 1
    
    return count
```

```python
added_mtl_pkgs_count = get_added_packages_mtl_by_lts(mtl_continuity_df)
removed_mtl_pkgs_count = get_removed_packages_mtl_by_lts(mtl_continuity_df)

mtl_count_continuity_df = pd.DataFrame(data={
    'lts': lts_list, 
    'added_mtl_pkgs_count': added_mtl_pkgs_count, 
    'removed_mtl_pkgs_count': removed_mtl_pkgs_count
})
```

```python
plt.figure(figsize=(18,6))
sns.lineplot(x='lts', y='value', hue='variable', data=pd.melt(mtl_count_continuity_df, ['lts']))

plt.savefig("graphs/added_removed_mlt_packages.png", bbox_inches='tight')
plt.show()
```

### Calcular el promedio de duracion de los paquetes en Stackage

```python
def sum_continuity(df):
    sum_list = list()
    
    for i, row in df.iterrows():
        sum = 0
        for j, lts in enumerate(lts_list):
            if row[lts] == 1:
                sum += 1
        sum_list.append(sum)
    return sum_list
    
continuity_df['lts_sum'] = sum_continuity(continuity_df)
```

### Analizar al continuidad de los paquetes en Stackage
- Cuantos paquetes se mantuvieron en todas las verisones LTS
- Cuantos paquetes fueron quitados y volvieron en futuras versiones LTS

```python
# retorna una lista de paquetes que estuvieron presentes en todos las versiones LTS
def get_full_continuity_packages(df):
    packages = list()
    for i, row in df.iterrows():
        count = 0;
        for j, lts in enumerate(lts_list):
            if(not df_list[j].loc[df_list[j]['package'].isin([row.name])].empty):
                count += 1
        if(count == len(lts_list)):
            packages.append(row.name)
    return packages

full_continuity_pkgs = get_full_continuity_packages(continuity_df)
```

```python
def get_packages_that_returned(df):
    packages = list()
    for i, row in df.iterrows():
        flag = False
        idx = 0;
        for j, lts in enumerate(lts_list):
            if(not df_list[j].loc[df_list[j]['package'].isin([row.name])].empty):
                if(idx == 0):
                    idx = j
                elif(j-idx > 1):
                    flag = True
                else:
                    idx = j

        if(flag):
            packages.append(row.name)
    return packages

pkgs_that_returned = get_packages_that_returned(continuity_df)
```

### Identificar cuantas veces cambio de version cada paquete para tener un indice de medicion en la mentencion

```python
def build_version_matrix(index):
    df = pd.DataFrame(index=index, columns=lts_list)

    for i, row in df.iterrows():
        for j, lts in enumerate(lts_list):
            version = 0
            package = df_list[j].loc[df_list[j]['package'].isin([row.name])]

            if(not package.empty):
                [version] = package['version'].values
            df.at[row.name, lts] = str(version)
            
    return df

version_df = build_version_matrix(all_time_packages)
```

# MTL-Direct usage

```python
# definir bien la lista de uso de monadas / mtl vs no-mtl

mtl_modules = [
    "Control.Monad.Cont", 
    "Control.Monad.Cont.Class", 
    "Control.Monad.Error",
    "Control.Monad.Error.Class",
    "Control.Monad.Except",
    "Control.Monad.Identity", 
    "Control.Monad.List", 
    "Control.Monad.RWS", 
    "Control.Monad.RWS.Class", 
    "Control.Monad.RWS.Lazy", 
    "Control.Monad.RWS.Strict", 
    "Control.Monad.Reader", 
    "Control.Monad.Reader.Class",
    "Control.Monad.State",
    "Control.Monad.State.Class",
    "Control.Monad.State.Lazy", 
    "Control.Monad.State.Strict", 
    "Control.Monad.Trans", 
    "Control.Monad.Writer", 
    "Control.Monad.Writer.Class", 
    "Control.Monad.Writer.Lazy", 
    "Control.Monad.Writer.Strict"
]

mtl_monads = ['Continuation', 'Error', 'Except', 'Identity', 'List', 'RWS', 'Reader', 'State', 'Trans', 'Writer']
```

```python
def usageCombinationToString(usage_vector):
    letters = ('C', 'E', 'I', 'L', 'R', 'S', 'T', 'W', 'X', 'Z')    
    combStr = "".join(list(map(lambda x: x[1] if x[0] == 1 else "", zip(usage_vector, letters))))
    if combStr == "":
        combStr = "None"
    return combStr
```

```python
def computeMonadUsageForDataframe(theDf):
    mtl_cont = mtl_modules[0:2]
    mtl_error = mtl_modules[2:4]
    mtl_except = mtl_modules[4:5]
    mtl_identity= mtl_modules[5:6]
    mtl_list = mtl_modules[6:7]
    mtl_rws = mtl_modules[7:11]
    mtl_reader = mtl_modules[11:13]
    mtl_state = mtl_modules[13:17]
    mtl_trans = mtl_modules[17:18]
    mtl_writer = mtl_modules[18:]    
    
    theDf.loc[theDf.index, 'Continuation']  = theDf[mtl_cont].sum(axis=1).apply(lambda x: 1 if x > 0 else 0)
    theDf.loc[theDf.index, 'Error']         = theDf[mtl_error].sum(axis=1).apply(lambda x: 1 if x > 0 else 0)
    theDf.loc[theDf.index, 'Except']        = theDf[mtl_except].sum(axis=1).apply(lambda x: 1 if x > 0 else 0)
    theDf.loc[theDf.index, 'Identity']      = theDf[mtl_identity].sum(axis=1).apply(lambda x: 1 if x > 0 else 0)
    theDf.loc[theDf.index, 'List']          = theDf[mtl_list].sum(axis=1).apply(lambda x: 1 if x > 0 else 0)
    theDf.loc[theDf.index, 'RWS']           = theDf[mtl_rws].sum(axis=1).apply(lambda x: 1 if x > 0 else 0)
    theDf.loc[theDf.index, 'Reader']        = theDf[mtl_reader].sum(axis=1).apply(lambda x: 1 if x > 0 else 0)
    theDf.loc[theDf.index, 'State']         = theDf[mtl_state].sum(axis=1).apply(lambda x: 1 if x > 0 else 0)
    theDf.loc[theDf.index, 'Trans']         = theDf[mtl_trans].sum(axis=1).apply(lambda x: 1 if x > 0 else 0)
    theDf.loc[theDf.index, 'Writer']        = theDf[mtl_writer].sum(axis=1).apply(lambda x: 1 if x > 0 else 0)
    theDf.loc[theDf.index, 'MonadsUsed']    = theDf[mtl_monads].sum(axis=1)
    
    theDf.loc[theDf.index, 'MonadsUsedVector'] = theDf.apply(lambda row: (
                                                                    row["Continuation"],
                                                                    row["Error"],
                                                                    row["Identity"],
                                                                    row["List"],
                                                                    row["Reader"],
                                                                    row["State"],
                                                                    row["Trans"],
                                                                    row["Writer"],
                                                                    row["Except"],
                                                                    row["RWS"]
                                                                ), axis = 1)
    
    theDf.loc[theDf.index, 'MonadsUsedCode'] = theDf.apply(lambda row: usageCombinationToString(row['MonadsUsedVector']), axis = 1)
    
    return theDf
```

```python
monads_used = [0, 1, 2, 3, 4, 5, 6, 7, 8]
data = create_lts_obj()
monads_used_df = []

for idx, df in enumerate(df_list):
    df = df[df['mtl-direct'] == 1]
    df = computeMonadUsageForDataframe(df)
    aux_df = df[['MonadsUsed', 'package']]
    aux_df = aux_df.groupby(['MonadsUsed']).count().reset_index().rename(columns={'package': 'count'})
    monads_used_df.append(aux_df)
    
for idx, lts in enumerate(data):
    total_pkgs_mtl = len(df_list[idx][df_list[idx]['mtl-direct'] == 1])
    
    for num in monads_used:
        count = monads_used_df[idx][monads_used_df[idx]['MonadsUsed'] == num]['count']
        if(count.empty):
            count = 0
        percentage = int(count) * 100 / total_pkgs_mtl
        data[lts].append(percentage)
    
monads_df = pd.DataFrame(data, index=monads_used)

ax = monads_df.plot.bar(figsize=(18,9), rot=0)
plt.savefig("graphs/count_monads_usage.png", bbox_inches='tight')
```

```python
# retorna una lista con todos los paquetes existentes en alguna version de LTS
def get_all_time_packages():
    all_packages = set()
    for df in df_list:
        lts_packages = set(df['package'])
        all_packages = all_packages.union(lts_packages)

    return list(all_packages)
```

```python
def get_current_top_used_packages(top):
    pkgs = {}
    
    for i, row in df_list[14].iterrows():
        for dependency in (list(set(row['deps']))):
            if(not dependency in pkgs):
                pkgs[dependency] = 1
            else:
                pkgs[dependency] += 1
                
    pkgs = sorted(pkgs, key=pkgs.get, reverse=True)
    return pkgs[:top]
```

```python
all_pkgs = get_all_time_packages()
top_used_pkgs = get_current_top_used_packages(100)
```

```python
##  packages that do not exist in stackage
def get_packages_out_of_stackage(pkgs):
    all_pkgs = get_all_time_packages()
    return list(filter(lambda pkg: pkg not in all_pkgs, pkgs))
```
